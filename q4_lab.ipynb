```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload the image
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

# Read and convert to grayscale
img = cv2.imread(img_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Show original grayscale image
print("Original Grayscale Image:")
cv2_imshow(gray)

# Apply CLAHE (local histogram equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
local_eq = clahe.apply(gray)

# Show CLAHE result
print("\nCLAHE (Local Histogram Equalized) Image:")
cv2_imshow(local_eq)

# Plot histograms
plt.figure(figsize=(12, 4))

# Original histogram
plt.subplot(1, 2, 1)
plt.hist(gray.ravel(), 256, [0, 256], color='gray')
plt.title("Original Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# CLAHE histogram
plt.subplot(1, 2, 2)
plt.hist(local_eq.ravel(), 256, [0, 256], color='green')
plt.title("Local Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
